```python
import paho.mqtt.client as mqtt
import time
import pandas as pd

broker = "test.mosquitto.org"
topic = "iiot/anomaly_detection/data"

# Charger ton dataset (ajuste le chemin si besoin)
train_df = pd.read_csv('/content/data/CMaps/train_FD001.txt', sep='\s+', header=None)

client = mqtt.Client()
client.connect(broker, 1883, 60)

for i in range(500):  # par exemple les 500 premières lignes seulement
    row = train_df.iloc[i].tolist()
    payload = ','.join([str(x) for x in row])
    client.publish(topic, payload)
    print(f"Message envoyé {i+1}/500")
    time.sleep(0.1)


client.disconnect()
```
